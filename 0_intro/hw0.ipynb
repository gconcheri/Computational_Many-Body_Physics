```python
X = np.array([[0, 1],[1, 0]])
Y = np.array([[0, -1j],[1j, 0]])
Z = np.array([[1, 0],[0, -1]])
```

```python
X, Y, Z
```

```python
H = np.kron(X, X)+np.kron(Y, Y)+np.kron(Z, Z)
H
```

```python
gs, E_gs = np.linalg.eigh(H)[1][:,0], np.linalg.eigh(H)[0][0]
print(gs)
print(E_gs)

ts, E_ts = np.linalg.eigh(H)[1][:,2], np.linalg.eigh(H)[0][2]
print(ts)
print(E_ts)
```

```python
# magnetization
M = np.kron(Z, np.eye(2))+np.kron(np.eye(2), Z)
M
```

```python
# grouns state
exp_M_g = np.dot(gs, np.dot(M, gs.reshape([4,1])))
exp_M_squared_g = np.dot(gs, np.dot(M**2, gs.reshape([4,1])))
print("expected magnetizazion on gs: ", exp_M_g)
M_var_g = exp_M_squared_g - exp_M_g**2
print("M variance: ", M_var_g)
```

```python
# second state
ts = np.array([1,0,0,1]/np.sqrt(2))
exp_M_t = np.dot(ts, np.dot(M, ts.reshape([4,1])))
exp_M_squared_t = np.dot(ts, np.dot(M**2, ts.reshape([4,1])))
print("expected magnetizazion on ts: ", exp_M_t)
M_var_t = exp_M_squared_t - exp_M_t**2
print("M variance: ", M_var_t)
```
